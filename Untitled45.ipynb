# CS987-Task 1: Gene Expression Prediction using Neural Networks

This notebook implements a deep learning pipeline to predict **lacY gene expression** using other gene features from *E. coli*. It includes **Exploratory Data Analysis (EDA)**, **Feature Selection**, **Neural Network Architecture**, **Hyperparameter Tuning**, **Model Training**, **Evaluation**, and **Biological Interpretation**.

---
### 1. Import Libraries

We begin by importing all necessary Python libraries for data handling, preprocessing, modeling, evaluation, and visualization.


```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import random
import os
import warnings
warnings.filterwarnings('ignore')
```

### 2. Set Random Seeds for Reproducibility

To ensure our model is reproducible, we explicitly set seeds for **Python**, **NumPy**, and **PyTorch**.


```python
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)

set_seed(42)
```

### 3. Load and Explore the Dataset

Next, we load the dataset and display basic information about it. We also visualize the distribution of the **lacY gene expression**.


```python
# Load the dataset
df = pd.read_csv("/content/Ecoli70_sample_data.csv")

# Display basic information and check the first few rows
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns.tolist())
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Visualize the distribution of lacY expression
sns.histplot(df["lacY"], kde=True)
plt.title("Distribution of lacY Expression")
plt.show()
```

### 4. Correlation Analysis

We will visualize the correlation heatmap of all genes and analyze the relationship between the genes and the **lacY** gene expression.


```python
# Correlation analysis
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

# Feature selection using correlation with lacY
corr = df.corr()["lacY"].sort_values(ascending=False)
print("Top 10 positively correlated genes with lacY:")
print(corr[1:11])

print("\nTop 10 negatively correlated genes with lacY:")
print(corr[-10:])
```

### 5. Feature Selection

We use **LassoCV** and **Random Forest Regressor** to select the most important features for predicting **lacY** expression.


```python
# Feature selection using LassoCV
X = df.drop(columns=["lacY"])
y = df["lacY"]

# Lasso feature selection
lasso = LassoCV(cv=5, random_state=42, max_iter=10000)
lasso.fit(X, y)
lasso_selected = X.columns[lasso.coef_ != 0]
print(f"Lasso selected {len(lasso_selected)} features")

# Feature selection using Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)
importances = rf.feature_importances_
rf_selected = pd.Series(importances, index=X.columns).sort_values(ascending=False).head(20).index
print(f"Random Forest selected top 20 features")

# Combining biological and statistical features
final_selected_features = list(set(lasso_selected) | set(rf_selected))

# Filter the dataset with selected features
X_selected = X[final_selected_features]
```

### 6. Data Preprocessing

We standardize the selected features and split the data into training and testing sets.


```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"\nTraining set size: {X_train.shape}, Test set size: {X_test.shape}")

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)
```

### 7. Neural Network Architecture

We define the architecture of the neural network with two hidden layers. The model will use **ReLU activation** and **Dropout** for regularization.


```python
# Define the neural network architecture
class GeneExpressionNN(nn.Module):
    def __init__(self, input_dim, hidden_dims=[128, 64], dropout=0.3):
        super(GeneExpressionNN, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, hidden_dims[0]),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dims[0], hidden_dims[1]),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dims[1], 1)
        )

    def forward(self, x):
        return self.network(x)

# Instantiate the model
model = GeneExpressionNN(input_dim=X_train.shape[1])

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

### 8. Hyperparameter Tuning (Optional)

We can use **GridSearchCV** to optimize hyperparameters like learning rate and batch size. However, for neural networks, hyperparameter tuning is often performed manually.


```python
# Experiment with different learning rates and batch sizes
learning_rates = [0.001, 0.0001]
batch_sizes = [32, 64, 128]

# Loop through combinations of learning rate and batch size for manual tuning
for lr in learning_rates:
    for batch_size in batch_sizes:
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        print(f"Training with learning rate: {lr} and batch size: {batch_size}")
        # Add your training loop here using the batch size and learning rate
        # Evaluate the model performance (e.g., accuracy or MSE)
```

### 9. Train the Model

We train the neural network for 100 epochs and track the loss during training.


```python
# Train the model
num_epochs = 100
train_losses = []
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    # Forward pass
    y_pred = model(X_train_tensor)

    # Calculate the loss
    loss = criterion(y_pred, y_train_tensor)
    train_losses.append(loss.item())

    # Backward pass and optimization
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Plot the training loss curve
plt.plot(train_losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

### 10. Model Evaluation

We evaluate the model using **R**, **MSE**, and **MAE**. Additionally, we visualize the predicted lacY expression versus the actual lacY expression.


```python
# Evaluate the model
model.eval()
with torch.no_grad():
    y_pred = model(X_test_tensor)
    test_loss = criterion(y_pred, y_test_tensor)
    test_r2 = r2_score(y_test_tensor.numpy(), y_pred.numpy())
    test_mse = mean_squared_error(y_test_tensor.numpy(), y_pred.numpy())
    test_mae = mean_absolute_error(y_test_tensor.numpy(), y_pred.numpy())

print(f"Test Loss: {test_loss.item():.4f}")
print(f"Test R: {test_r2:.4f}")
print(f"Test MSE: {test_mse:.4f}")
print(f"Test MAE: {test_mae:.4f}")

# Visualize actual vs predicted lacY expression
plt.scatter(y_test, y_pred.numpy(), alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual lacY Expression')
plt.ylabel('Predicted lacY Expression')
plt.title('Actual vs Predicted lacY Expression')
plt.show()
```
